```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
df = pd.read_csv("Train.csv")
```

```python
f1 = np.asarray(list(df.get('feature_1')))
```

```python
f2 = np.asarray(list(df.get('feature_2')))
f3 = np.asarray(list(df.get('feature_3')))
f4 = np.asarray(list(df.get('feature_4')))
f5 = np.asarray(list(df.get('feature_5')))
Y = np.asarray(list(df.get('target')))
print(Y)
```

```python
def hypothesis(theta,f1,f2,f3,f4,f5):
    return theta[0]+theta[1]*f1+theta[2]*f2+theta[3]*f3+theta[4]*f4+theta[5]*f5
```

```python
def gradient(theta,f1,f2,f3,f4,f5,Y):
    grad = np.zeros((6))
    m = Y.shape[0]
    
    for i in range(m):
        hx = hypothesis(theta,f1[i],f2[i],f3[i],f4[i],f5[i])
        grad[0]+= (hx-Y[i])
        grad[1]+= (hx-Y[i])*f1[i]
        grad[2]+= (hx-Y[i])*f2[i]
        grad[3]+= (hx-Y[i])*f3[i]
        grad[4]+= (hx-Y[i])*f4[i]
        grad[5]+= (hx-Y[i])*f5[i]
    return grad    
        
```

```python
def gradient_descent(theta,f1,f2,f3,f4,f5,Y,lr=0.0001,max_itr = 100):
   
    for i in range (max_itr):
        grad = gradient(theta,f1,f2,f3,f4,f5,Y)
        theta[0] = theta[0]-lr*grad[0]
        theta[1] = theta[1]-lr*grad[1]
        theta[2] = theta[2]-lr*grad[2]
        theta[3] = theta[3]-lr*grad[3]
        theta[4] = theta[4]-lr*grad[4]
        theta[5] = theta[5]-lr*grad[5]
    return theta
```

```python
theta = np.zeros((6))
```

```python
theta = gradient_descent(theta,f1,f2,f3,f4,f5,Y)
```

```python
df  = pd.read_csv("Test.csv")
T1 = np.asarray(list(df.get('feature_1')))
T2 = np.asarray(list(df.get('feature_2')))
T3 = np.asarray(list(df.get('feature_3')))
T4 = np.asarray(list(df.get('feature_4')))
T5 = np.asarray(list(df.get('feature_5')))
y = list()
for i in range(400):
    a =  hypothesis(theta,T1[i],T2[i],T3[i],T4[i],T5[i])
    y.append(a)
print(y)    
```

```python
df = pd.DataFrame(y)
df.to_csv("submission.csv")
```
